## Clean Data


This dataset contains information about 891 people who were on board the ship when departed on April 15th, 1912. Some people aboard the ship were more likely to survive the wreck than others. The challenge is to build a model to predict which people would survive based on the following fields:

- **Name** (str) - Name of the passenger
- **Pclass** (int) - Ticket class
- **Sex** (str) - Sex of the passenger
- **Age** (float) - Age in years
- **SibSp** (int) - Number of siblings and spouses aboard
- **Parch** (int) - Number of parents and children aboard
- **Ticket** (str) - Ticket number
- **Fare** (float) - Passenger fare
- **Cabin** (str) - Cabin number
- **Embarked** (str) - Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)

### Read in Data

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
df = pd.read_csv('Data/titanic.csv')
df.head()
```

### Clean continuous variables

#### Fill missing for `Age`

```python
df.isnull().sum()
```

```python
df.Age.fillna(df.Age.mean(), inplace=True)
df.head(10)
```

#### Combine `SibSp` & `Parch`

```python
for i, col in enumerate(['SibSp', 'Parch']):
    plt.figure(i)
    sns.catplot(x=col, y='Survived', data=df, kind='point', aspect=2)
```

```python
df['Family_count'] = df.SibSp + df.Parch
```

#### Drop unnnecessary variables

```python
df.drop(columns = ['PassengerId', 'SibSp', 'Parch'], inplace=True)
```

```python
df.head()
```

### Clean categorical variables

#### Fill in missing & create indicator for `Cabin`

```python
df.isnull().sum()
```

```python
df.groupby(df.Cabin.isnull())['Survived'].mean()
```

```python
df['Cabin_ind'] = np.where(df.Cabin.isnull(), 0, 1)
df.head()
```

#### Convert `Sex` to numeric

```python
df.Sex.replace('male', 0, inplace = True)
df.Sex.replace('female', 1, inplace = True)
```

```python
df.head()
```

#### Drop unnecessary variables

```python
df.drop(['Cabin', 'Embarked', 'Name', 'Ticket'], axis=1, inplace=True)
df.head()
```

### Write out cleaned data

```python
df.to_csv('Data/titanic_cleaned.csv', index=False)
```
